```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

%matplotlib inline
```

## Task: Improve sunspot number predictions on the basis of combined method by assimilating the 6 last available monthly mean sunspot numbers. 

```python
# data
sunspot_number = np.loadtxt("data/sunspot_number.dat")
sunspot_number_ckeck = np.loadtxt("data/sunspot_number_check.dat")
SN_predictions = np.loadtxt("data/SN_predictions.txt")
```

```python
SN_predictions[:,0]
```

```python
sunspot_number_ckeck
```

```python
plt.figure(figsize=(17,12))
n0 = 3000
ckeck_interval = np.arange(-59, -1)
ckeck_smooth_interval = np.arange(-67, -7)
plt.plot(sunspot_number[n0:,1], sunspot_number[n0:,2], label="Monthly mean sunspot number")
plt.plot(sunspot_number[n0:-7,1], sunspot_number[n0:-7,3], label="13-month running mean")
plt.plot(SN_predictions[:,0] + (SN_predictions[:,1] - 1) / 12., SN_predictions[:,2], label="Prediction")
plt.plot(sunspot_number_ckeck[ckeck_interval,1], sunspot_number_ckeck[ckeck_interval,2], "--", label="Unnown data")
plt.plot(sunspot_number_ckeck[ckeck_smooth_interval,1], sunspot_number_ckeck[ckeck_smooth_interval,3], "--", label="Unnown smoothed data")
plt.legend()
plt.grid()
plt.show()
```
